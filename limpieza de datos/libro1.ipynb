# **Limpieza de datos**

```python
import pandas as pd
df= pd.read_excel('data_frame.xlsx')
```

```python
import re
from unicodedata import normalize

def normalizarP(Programa_Academico):
    palabra=re.sub(r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", 
            normalize("NFD",Programa_Academico), 0, re.I)
    palabra=normalize('NFC',palabra)
    return palabra

def cambiarPalabra(programa):
    if "INGENIERA" in programa:
        return programa.replace("INGENIERA","INGENIERIA")
    else:
        return programa    
        
    
df['Programa Académico']=df['Programa Académico'].apply(normalizarP)
df['Programa Académico']=df['Programa Académico'].apply(cambiarPalabra)

df[(df['Programa Académico']=='INGENIERIA TELEMATICA')]
```

```python
def cambiarP(valor):
    if 'INGENIERIA EN REDES DE COMPUTACION' in valor:
        return 'INGENIERIA EN TELEMATICA'
    else:
        return valor

len(df[(df['Programa Académico'].str.contains('INGENIERIA EN REDES DE COMPUTACION'))])

df['Programa Académico']=df['Programa Académico'].apply(cambiarP)

len(df[(df['Programa Académico'].str.contains('INGENIERIA EN REDES DE COMPUTACION'))])

```

```python
auxiliar = df.groupby(['Programa Académico']).agg({
    "Código de la Institución":"max"
})

len(auxiliar)
```
